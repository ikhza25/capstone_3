### **Success Criteria:**

#### **Metric Evaluation**
Dengan fokus proyek kami adalah mengidentifikasi pelanggan yang kemungkinan akan membatalkan pemesanan hotel, kelas target kami didefinisikan sebagai berikut:

**Target**:

- 0 : Tidak terjadi pembatalan pemesanan hotel
- 1 : Terjadi pembatalan pemesanan hotel

#### **False Prediction :**  
Dalam konteks pemodelan machine learning untuk memprediksi pembatalan pemesanan kamar di hotel, evaluasi metrik membantu mengukur kinerja model dan dampaknya terhadap hotel.

| Actual \ Predicted | No Canceled (0)               | Canceled (1)                  |
|-------------------|--------------------------------|--------------------------------|
| **No Canceled (0)** | True Negative (TN)             | **False Positive (FP)**            |
|                   | Prediction: No Canceled       | **Prediction: Canceled**          |
|                   | Actual: No Canceled           | **Actual: No Canceled**           |
| **Canceled (1)**    | **False Negative (FN)**            | True Positive (TP)             |
|                   | **Prediction: No Canceled**       | Prediction: Canceled          |
|                   | **Actual: Canceled**              | Actual: Canceled              |

#### **Type 1 Error (False Positive)**

**Definisi**: Kesalahan ini terjadi ketika model memprediksi bahwa pemesanan akan dibatalkan, padahal kenyataannya tidak terjadi pembatalan.

**Konsekuensi**:

Jika model mengidentifikasi pemesanan sebagai berisiko tinggi untuk dibatalkan dan hotel mengalihkan kamar tersebut ke pelanggan lain, pelanggan yang sebenarnya tidak membatalkan mungkin merasa tidak dihargai atau kecewa. Ini dapat menyebabkan hilangnya pelanggan loyal dan dampak negatif pada reputasi hotel, serta kerugian pendapatan yang mungkin timbul dari pelanggan yang tidak puas. Misalkan dengan asumsi:

- Biaya Operasional:
    Persiapan kamar untuk rebooking mungkin memerlukan biaya tambahan. Misalnya, biaya persiapan kamar adalah **€40,00**.

- Rata-rata Tarif Kamar (ARR):
    Mengacu pada rata-rata biaya akomodasi di Portugal sebesar $125 per malam, yang dikonversi menjadi **€112,50 (dengan kurs 1 USD = 0,90 EUR)**  berdasarkan situs di Travellers Worldwide.

- Probabilitas Penghuni Kamar:
    Probabilitas kamar terisi jika tidak ada pembatalan adalah **65%**.

- Biaya Akuisisi Pelanggan (CAC):
    Jika pelanggan yang hilang perlu diganti, biaya akuisisi pelanggan bisa mencapai **€80,00**.

- Kehilangan Nilai Seumur Hidup Pelanggan (CLV):
    Jika pelanggan yang tidak puas tidak kembali, potensi kehilangan nilai seumur hidup pelanggan adalah **€400,00**.

#### **Perhitungan Total Biaya**:

**Total Loss Cost = Operational Cost + (Average Room Rate (ARR) × Occupancy Probability) + Customer Acquisition Cost (CAC) + Customer Lifetime Value (CLV) Loss**

Substitusi nilai-nilai tersebut:</br>
**Total Loss Cost = €40,00 + (€112,50 × 0,65) + €80,00 + €400,00** </br>
**Total Loss Cost = €40,00 + €73,13 + €80,00 + €400,00**</br>
**Total Loss Cost = €593,13**

#### **Type 2 Error (False Negative)**

**Definisi**: Kesalahan ini terjadi ketika model memprediksi bahwa pemesanan tidak akan dibatalkan, padahal kenyataannya pembatalan terjadi.

**Konsekuensi**:

Hotel mungkin tidak siap menghadapi pembatalan yang tidak terduga dan kehilangan kesempatan untuk mengisi kamar kosong dengan pelanggan baru, terutama jika pembatalan terjadi mendekati waktu check-in. Hal ini mengakibatkan pendapatan yang hilang karena kamar kosong yang tidak terisi. **Misalkan dengan asumsi**

- Rata-rata Tarif Kamar (ARR):
    Mengacu pada rata-rata biaya akomodasi di Portugal sebesar $125 per malam, yang dikonversi menjadi **€112,50 (dengan kurs 1 USD = 0,90 EUR)** berdasarkan situs Travellers Worldwide.

- Probabilitas Penghuni Kamar:
    Probabilitas kamar terisi jika tidak ada pembatalan adalah **65%**.

- Biaya Kompensasi:
    Jika hotel melakukan overbooking untuk mengantisipasi pembatalan dan harus memberikan kompensasi kepada pelanggan yang dipindahkan, biaya kompensasi bisa mencapai **€45,00**.

Perhitungan Potensi Kerugian Pendapatan:

**Potensi Kerugian Pendapatan = Rata-rata Tarif Kamar (ARR) × Probabilitas Penghuni Kamar**

Substitusi nilai-nilai tersebut:
- **Potensi Kerugian Pendapatan = €112,50 × 0,65**
- **Potensi Kerugian Pendapatan = €73,13**

**Total Kerugian = Potensi Kerugian Pendapatan + Biaya Kompensasi**

Substitusi nilai-nilai tersebut:
- **Total Kerugian = €73,13 + €45,00**
- **Total Kerugian = €118,13**

```python
# Feature Creation Classes
class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X = pd.DataFrame(X)
        X['HasDeposit'] = X['deposit_type'].apply(lambda x: 1 if x != 'No Deposit' else 0)
        X['IsOnlineBooking'] = X['market_segment'].apply(lambda x: 1 if x in ['Online TA', 'Direct'] else 0)
        X['TotalBookingChangesAndRequests'] = X['booking_changes'] + X['total_of_special_requests']
        return X
```
